**Hint:** `random.choice()` 

A function called `random.choice()` takes a list and returns a random value from it. You can see how this function works in the example below. This might come in useful!  

```python
from random import choice

# run this code a couple of times to see what happens
some_list = ['a', 'b', 'c', 'd', 'e', 'f',]
print(choice(some_list))
```

### Write your Task 2 code below:

```python
# your Task 2 code here. You can use multiple cells if you wish. 
from random import choice

score = 0
numofguesses = 0
for item in countries:
    while numofguesses < 10:
        random_country = choice(countries)
        guess_country = input(random_country['capital'] + " is the capital of? ")
        numofguesses += 1
        if guess_country == random_country['name']:
            score += 1
            print("Correct answer! ")
        else:
            print("Wrong answer!")
        print()
        break
print("You got " + str(score)+"/10")
```

I began by ensuring the random choice function is enabled.\

I then put in a for loop to refer back to the list of countries, and put a while loop under it to limit the numofguesses to 10. 
under the while loop, I defined a variable called "random_country" which will present a random choice to the user, and also another variable called "guess_country" which is the part that gives the question and input to the user.
I also put "numofguesses += 1" so that with every guess, it increase by 1 until it reaches 10 questions. 
I then utilised the if / else statements to differentiate the response based on the correctness of the answer. If the answer is correct, "score" will increase by one and the program also tells the user that they have guessed correctly. 
The else statement is used to print out "Wrong answer!" if the answer was incorrect. 

finally, the program prints out a message with the total score.
